__Dictionary for room type__
1. Entire home/apt  - 1
2. Private room     - 2
3. Shared room      - 3
4. Hotel room       - 4

```python
data['room_type'] = data['room_type'].apply(lambda x : 1 if x == "Entire home/apt" else 2 if x == "Private room" else 3 if x == "Shared room"  else 4 if x == "Hotel room" else 0)
```

```python
data['cancellation_policy'].value_counts()
```

__Dictionary for contract policy__
1. moderate  - 1
2. strict    - 2
3. flexible  - 3


```python
#data['cancellation_policy'] = data['cancellation_policy'].replace("moderate",1)
```

```python
type(data['cancellation_policy'][1])
```

```python
data['cancellation_policy'] = data['cancellation_policy'].apply(lambda x : 2 if x == "strict" else 3 if x == "flexible" else 1 if x == "moderate" else 0)
```

```python
data['host_identity_verified'].value_counts()
```

__Dictionary for host_identity_verifies__
1. Unconfirmed    -1
2. verified       -2

```python
data['host_identity_verified'] = data['host_identity_verified'].apply(lambda x: 1 if x == "UNCONFIRMED" else 2 if x == "VERIFIED" else 0)
```
