# 4. Cleaning individual columns

```python
data
```

```python
data['rating'] = data['rating'].str.replace("/5","")
```

```python
data['cost_for_2_people'] = data['cost_for_2_people'].str.replace(",","") 
```

```python
data['online_order'].value_counts()
```

```python
data['online_order'] = data['online_order'].apply(lambda x: 1 if x=="Yes" else 2 if x == "No" else 0)
```

```python
data['table_booking_rate'].value_counts()
```

```python
data['table_booking_rate'] = data['table_booking_rate'].apply(lambda x: 1 if x=="Yes" else 2 if x == "No" else 0)
```

```python
data['service_type'].value_counts()
```

```python
data['service_type'] = data['service_type'].apply(lambda x: 1 if x=="Delivery" else 2 if x == "Dine-out" else 3 if x == "Desserts" else 4 if x == "Cafes" else 5 if x == "Drinks & nightlife" else 6 if x == "Buffet" else 7 if x == "Pubs and bars" else 0)
```
